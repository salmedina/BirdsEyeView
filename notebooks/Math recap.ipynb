```python
%matplotlib inline
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import open3d as o3
from open3d import JVisualizer
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
r = 6
def f(x, y):
    return -np.sqrt(r**2 - x**2 - y**2)

x = np.linspace(-r, r, 30)
y = np.linspace(-r, r, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
